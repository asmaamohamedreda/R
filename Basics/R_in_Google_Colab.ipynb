```r
df = read.table("/content/sample_data/expData.txt")
head(df, n=10)
```

```r
avg.exp = rowMeans(df)
avg.exp
```

```r
df = cbind(df, avg.exp)
head(df, n= 10)
```

```r
dim(df)
```

```r
# Apply function: 1 (Rows) , 2(columns)
mean_exp = apply(df, 1, mean)
df = cbind(df, mean_exp)
```

```r
head(df, n= 10)
```

```r
# Max expression for each gene
max_exp = apply(df, 1, max)
df = cbind(df, max_exp)
head(df, n=10)
```

```r
# min, sum, sd
min_Exp = apply(df, 1, min)
sum_exp = apply(df, 1, sum)
sd_Exp = apply(df, 1, sd)

df = cbind(df, min_Exp, sum_exp,sd_Exp )
head(df, n= 10)
```

```r
# quantile expression:
qunt_exp = apply(df, 1, quantile, na.rm=T)
qunt_exp
```

```r
df = cbind(df, t(qunt_exp))
head(df, n= 10)
```

```r
specificQuantile2080 = apply(df, 1, quantile, probs = c(0.2, 0.8), na.rm= T)
specificQuantile2080
```

```r
df = cbind(df, t(specificQuantile2080))
head(df, n=10)
```
